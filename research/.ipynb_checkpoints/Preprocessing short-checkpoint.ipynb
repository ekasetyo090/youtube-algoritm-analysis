## Transform Views Data
### Transform Function

```python
def view_to_numeric(x):
    identifier = {'K':1000,'M':1000000}
    spliter = x.split(' views')
    if spliter[0][-1] in identifier.keys():
        num  = float(spliter[0][:-1])*identifier.get(spliter[0][-1])
        return(round(num))
    else: 
        return x
```

### Transforming Data

```python
df['short_view_count'] = df['short_view_count'].apply(lambda row: view_to_numeric(row))
```

## Spliting Headline Per Word

```python
df['short_headline'] = df['short_headline'].apply(lambda row: row.lower().split(' '))
```
